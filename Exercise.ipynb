# 9. Nan Values: 

```python
#checking nan values in the dataframe
print(df.isnull().sum())
```

```python
df.PetalLengthCm[62]
```

```python
nan_dict = {}

all_columns =  ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm', 'Species']
for column in df[all_columns]:
     if df[column].isna().any():
        # Store the indices of missing values
        nan_indices = df[df[column].isna()].index.tolist()
        nan_dict[column] = nan_indices
        
         # Check if column is numerical or categorical
        if df[column].dtype == np.number: # Randle: this throws a deprecation warning. Maybe better to use something like float, int and/or str or some other check. It does seem to work though
            # Replace missing values with the mean of the column
            df[column].fillna(df[column].mean(), inplace=True)
        else:
            # Drop rows with missing values in categorical column
            df.dropna(subset=[column], inplace=True)

# Reset the row index
df = df.reset_index(drop=True)

# Print the updated DataFrame
print(df)

# Print the nan_dict
print(nan_dict)
```

# 10. Outliers:

```python
 # Calculate mean and standard deviation
 mean = df['SepalWidthCm'].mean()
 std = df['SepalWidthCm'].std()

 # Calculate the value/score that is 2 standard deviation above the mean
 value = mean + 2 * std

# Randle: careful with indentation: the above lines is slightly indented relative to the other text

# Find row indices with values outside the range
indices = df.loc[df.SepalWidthCm >= value].index.tolist() # Randle: not quite correct, this will only detect outliers to the right of the mean.

# Drop rows with identified indices
df = df.drop(indices)

# Reset the row index
df = df.reset_index(drop=True)

df.shape
```

# 13. Normalize numerical columns: 

```python
# Create copies of the original DataFrame
df_minmax = df.copy()
df_z = df.copy()

# apply normalization techniques

numerical_columns = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
for column in df[numerical_columns]:
   df_minmax [column] = (df_minmax[column] - df_minmax[column].min()) / (df_minmax[column].max() - df_minmax[column].min())    
  
# view normalized data
print(df_minmax) # Randle: to be sure that it worked, its good to use .describe() again

# apply normalization techniques
for column in df[numerical_columns]:
  df_z[column] = (df_z[column] - df_z[column].mean()) / df_z[column].std()    
  
# view normalized data   
display(df_z) # Randle: to be sure that it worked, its good to use .describe() again
```
